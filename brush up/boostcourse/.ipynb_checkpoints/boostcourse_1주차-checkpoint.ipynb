```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/healthexp.csv")
df
```

```python
df.loc[df['Year']>=2011].pivot_table("Life_Expectancy","Year","Country",aggfunc='mean',margins=True, fill_value=0)
```

```python
df.loc[df['Year']>=2011].pivot_table(values=["Life_Expectancy"], index=["Year"], columns=["Country"],aggfunc='mean', margins=True, fill_value=0)
```

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/healthexp.csv")
df.loc[df['Year']>=2011].pivot_table("Life_Expectancy", "Year", "Country", aggfunc='mean', margins=True)
```

### 1.판다스란?
> 데이터 조작 및 분석을 위한 Python 프로그래밍 언어용으로 작성된 소프트웨어 라이브러리    
> 행과 열로 이루어진 데이터 객체를 만들어 다룰수 있게 되며 안정적으로 대용량 데이터들을 처리하는데 유용함   

### 2. 판다스 사용하기   
- 판다스 설치하기 : 아나콘다 cmd에서 작성 가능 (개인 가상환경이 있다면 activate로 활성화 시킨 후 설치 가능)    
    * pip install pandas   
    
- python코드에 import하기   
    * import pands as pd    
    * 'as pd' : Alias(별칭)약자인 as를 사용하여 pandas를 pd인 별칭으로 사용 (단순 편의기능)   
    
### 3. 판다스 간단한 문법   
* Pandas를 통한 파일 저장과 불러오기   
    * to_csv("파일명", index=False) : csv 파일로 저장하기   
    * read_csv("파일명") : csv 파일 불러오기   
    * shape를 통한 행과 열의 수 보기   
    * head, tail, sample을 통한 일부 데이터 가져오기   
    
* DataFrame의 info(), describe()등을 통한 요약과 기술통계 값 구하기   
    * info()   
    * describe()   
    * nunique()   
    * index   
    * columns   
    * values   
    
* Pandas의 DataFrame과 Series의 이해   
    * Series : 1차원 벡터구조   
    * DataFrame : 2차원 행렬구조   
    
* 색인하기   
    * [컬럼]   
    * .loc[행]   
    * .loc[행, 열]   
    * .loc[조건식, 열]   

* DataFrame의 데이터 타입 이해하기   
    * 날짜 데이터의 변환   
    
* DataFrame 다루기   
    * column indexing   
    * index indexing   
    * column & index indexing   
    * sort : use to 'sort_values'
    * 조건식 사용하기   
    
* 빈도수 구하기   
    * 한 개의 변수 : series.value_counts()   
    * 두 개의 변수 : pd.crosstab()   
    
* groupby와 pivot_table   
    * 다양한 연산식의 사용

### 4. 간단한 예제를 통한 문법 적용
