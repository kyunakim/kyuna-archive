### Q4. 평균 기대수명이 점점 길어지면서 실버 의료 산업도 주목받고 있습니다. 여러분은 실버 의료 산업과 관련된 스타트업에 취업했습니다. 지도를 시각화하여 '노인/치매병원'이 주로 어디에 위치하고 있는지를 찾아보고자 합니다.

- folium 을 통해 지도에 전국의 '노인/치매병원'을 표시해 주세요!
- 다음과 같이 시각화 하며, folium 의 문서를 참고하여 다른 기능을 사용하여 좀 더 멋지게 지도를 꾸며도 좋습니다.
- folium 문서 : https://python-visualization.github.io/folium/quickstart.html

```python
import folium
```

```python
df_noin = df[df['상권업종소분류명']=='노인/치매병원'].copy()
df_noin.head(3)
```

```python
# 지도의 중심을 지정하기 위해 위도와 경도의 평균을 구합니다. 
fmap = folium.Map(location=[df_noin['위도'].mean(), df_noin['경도'].mean()], zoom_start=12)

for n in df_noin.index:
    # 팝업에 들어갈 텍스트를 지정
    # 상호명과 주소를 붙여서 출력되도록 한다.
    popup_name = df_noin.loc[n,'상호명'] + ' - ' + df_noin.loc[n, '도로명주소']
    
    folium.Marker(
        location=[df_noin.loc[n,'위도'], df_noin.loc[n,'경도']],
        popup=popup_name,
        tooltip=popup_name
    ).add_to(fmap)

fmap
```
